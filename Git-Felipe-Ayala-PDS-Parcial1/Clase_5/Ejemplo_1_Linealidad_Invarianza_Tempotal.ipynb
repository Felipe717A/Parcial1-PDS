# Ejemplo 1. Linealidad e Invarianza Temporal

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal de entrada
t=np.arange(0,100)
x = np.cos(50*t)

# Transformación lineal e invariante en el tiempo
T = lambda x: 2*x + 3

# Señal de salida
y = T(x)

# Prueba de linealidad
a = 2
b = 3
y1 = T(a*x + b)
y2 = a*T(x) + b
assert np.allclose(y1, y2)

# Prueba de invarianza temporal
t_shift = 2
y1 = T(x + t_shift)
y2 = T(x) + T(t_shift*np.ones_like(x))
#assert np.allclose(y1, y2)

# Gráfica de las señales
plt.plot(x, y, label='Señal de salida')
plt.plot(x, a*x + b, label='Señal de entrada')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal de entrada
x = np.sin(np.linspace(0, 2*np.pi, 100))

# Transformación lineal e invariante en el tiempo
T = lambda x: 2*x + 3

# Señal de salida
y = T(x)

# Prueba de linealidad
a = 2
b = 3
y1 = T(a*x + b)
y2 = a*T(x) + b
assert np.allclose(y1, y2)

# Prueba de invarianza temporal
t_shift = 10
y1 = T(np.roll(x, t_shift))
y2 = np.roll(T(x), t_shift)
assert np.allclose(y1, y2)

# Gráfica de las señales
plt.plot(x, label='Señal de entrada')
plt.plot(y, label='Señal de salida')
plt.legend()
plt.show()
```

```python
En este código, la señal de entrada x es un seno con 100 puntos que se extiende a lo largo de un período completo. La transformación lineal e invariante en el tiempo T sigue siendo la misma que en el ejemplo anterior: multiplicar la entrada por 2 y luego agregar 3. La señal de salida y se obtiene aplicando la transformación T a la señal de entrada x.

Para probar la linealidad, creamos dos señales de entrada escalando y desplazando la señal original, y aplicamos la transformación T a ambas señales de entrada. Luego, comprobamos si la señal de salida obtenida al escalar y desplazar la señal original es igual a la suma de las señales de salida obtenidas al aplicar la transformación a cada señal de entrada escalada y desplazada.

Para probar la invarianza temporal, creamos una señal de entrada desplazando la señal original en el tiempo mediante la función np.roll(), que desplaza los valores de la señal una cantidad específica de posiciones. Luego, aplicamos la transformación T a ambas señales de entrada y comprobamos si la señal de salida obtenida al desplazar la señal original en el tiempo es igual a la señal de salida obtenida al aplicar la transformación a la señal original y luego desplazarla en el tiempo.

Finalmente, graficamos la señal de entrada y la señal de salida para visualizar el efecto de la transformación T sobre la señal original.
```
