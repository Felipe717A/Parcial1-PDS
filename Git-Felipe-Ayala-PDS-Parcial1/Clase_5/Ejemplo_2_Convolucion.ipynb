# Ejemplo 2. Convolución de Señales.

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal de entrada
x = np.array([1, 2, 1, 2])

# Kernel
h = np.array([0.5, 0.5])

# Convolución
y = np.convolve(x, h)

# Gráficas
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Señal de entrada
ax1.stem(x, use_line_collection=True)
ax1.set_title('Señal de entrada')
ax1.set_xlabel('Muestras')
ax1.set_ylabel('Amplitud')
ax1.grid()

# Kernel
ax2.stem(h, use_line_collection=True)
ax2.set_title('Kernel')
ax2.set_xlabel('Muestras')
ax2.set_ylabel('Amplitud')
ax2.grid()

fig.tight_layout()
plt.show()

# Gráfica de la convolución
plt.stem(y, use_line_collection=True)
plt.title('Señal de salida')
plt.xlabel('Muestras')
plt.ylabel('Amplitud')
plt.grid()
plt.show()
```

```python
En este código, definimos una señal de entrada x y un kernel h. El kernel representa la respuesta al impulso de un sistema, es decir, cómo un sistema responde a un impulso unitario. En este caso, el kernel es simplemente una media móvil de dos puntos.

Para calcular la convolución, usamos la función np.convolve(). La convolución es una operación que toma dos señales y produce una tercera señal que representa cómo una señal afecta a otra. Matemáticamente, la convolución se define como la integral del producto de dos señales, pero en el caso discreto que estamos considerando, la convolución se calcula simplemente sumando productos de muestras de las señales.

Después de calcular la convolución, graficamos la señal de entrada, el kernel y la señal de salida. La señal de entrada y el kernel se grafican usando la función plt.stem(), que muestra cada muestra como un pequeño segmento vertical. La señal de salida se grafica de la misma manera.
```
