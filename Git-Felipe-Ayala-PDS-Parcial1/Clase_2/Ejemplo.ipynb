```python
import numpy as np
import matplotlib.pyplot as plt

# Crear señal sinusoidal
f_signal = 10 # frecuencia de la señal (en Hz)
t = np.linspace(0, 1, 1000)
signal = np.sin(2*np.pi*f_signal*t)

# Mostrar la señal sinusoidal
fig, ax = plt.subplots()
ax.plot(t, signal)
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.set_title('Señal original')

# Calcular frecuencia de Nyquist
f_nyquist = 2 * 100

# Calcular frecuencia de muestreo
f_sampling = f_nyquist * 2

# Muestrear la señal
n_samples = int(len(t) / (f_sampling / f_signal))
samples_idx = np.arange(n_samples) * int(f_sampling / f_signal)
signal_sampled = signal[samples_idx]

# Mostrar la señal muestreada
fig, ax = plt.subplots()
#ax.plot(t, signal, label='Original')
ax.stem(t[samples_idx], signal_sampled, label='Muestreada')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.set_title('Señal muestreada')
ax.legend()

plt.show()

# Calcular frecuencia de Nyquist
f_nyquist = 2 * 10

# Calcular frecuencia de muestreo
f_sampling = f_nyquist * 2

# Muestrear la señal
n_samples = int(len(t) / (f_sampling / f_signal))
samples_idx = np.arange(n_samples) * int(f_sampling / f_signal)
signal_sampled = signal[samples_idx]

# Mostrar la señal muestreada
fig, ax = plt.subplots()
#ax.plot(t, signal, label='Original')
ax.stem(t[samples_idx], signal_sampled, label='Muestreada')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('Amplitud')
ax.set_title('Señal muestreada')
ax.legend()

plt.show()
```

Incremento de muestras gracias a variar la frecuencia de muestreo (tasa de nyquist)
