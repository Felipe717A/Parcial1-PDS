# Ejemplo 1. Digitalización de señales analógicas.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 20) # Vector de tiempo
signal = np.exp(t)*np.sin(t)              # Señal analogica
```

```python
# Grafica de la señal analógica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal analógica')
plt.stem(t, signal, 'r', label = 'Señal digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
