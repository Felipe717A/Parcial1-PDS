# Ejemplo. Filtros de Media Movil.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Definir los parmetros de la seal
fs = 1000  # Frecuencia de muestreo
f = 50  # Frecuencia de la seal
t = np.arange(0, 1, 1/fs)  # Vector de tiempo

# Generar la seal de coseno con ruido
noise = 1.5*np.random.randn(len(t))  # Componente de ruido
signal = np.cos(2*np.pi*f*t) + noise  # Seal con ruido

# Aplicar filtro de media mvil
window_size = 11  # Tamao de la ventana del filtro
filtered_signal = savgol_filter(signal, window_size, 1)  # Aplicar filtro

# Graficar la seal original y la seal filtrada
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# Seal original
ax[0].plot(t, signal)
ax[0].set_xlabel('Tiempo [s]')
ax[0].set_ylabel('Amplitud')
ax[0].set_title('Seal de coseno con ruido')

# Seal filtrada
ax[1].plot(t, filtered_signal)
ax[1].set_xlabel('Tiempo [s]')
ax[1].set_ylabel('Amplitud')
ax[1].set_title('Seal de coseno filtrada')

plt.tight_layout()
plt.show()
```

Uso del filtro de media movil para eliminar el ruido de una seal, demarcar tendencias

La funcin savgol_filter de la librera Scipy en Python aplica un filtro de suavizado de Savitzky-Golay a una seal dada. Este filtro es un tipo de filtro de media mvil que se utiliza comnmente para suavizar seales que contienen ruido. A diferencia de un filtro de media mvil simple, el filtro de Savitzky-Golay utiliza un ajuste polinmico local para suavizar la seal, lo que permite una mejor preservacin de las caractersticas importantes de la seal, como los bordes y las transiciones rpidas.

La funcin savgol_filter acepta tres argumentos principales: la seal de entrada que se va a suavizar, el tamao de la ventana del filtro y el orden del ajuste polinmico. El tamao de la ventana del filtro (window_size) especifica el nmero de puntos que se utilizarn en cada ajuste polinmico local, mientras que el orden (polyorder) especifica el orden del polinomio que se ajustar a los puntos dentro de cada ventana.

La funcin devuelve la seal suavizada despus de aplicar el filtro de Savitzky-Golay. En resumen, la funcin savgol_filter es una herramienta til para reducir el ruido en seales y preservar sus caractersticas importantes.
