# Ejemplo. Filtros de Media Movil.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Definir los parámetros de la señal
fs = 1000  # Frecuencia de muestreo
f = 50  # Frecuencia de la señal
t = np.arange(0, 1, 1/fs)  # Vector de tiempo

# Generar la señal de coseno con ruido
noise = 1.5*np.random.randn(len(t))  # Componente de ruido
signal = np.cos(2*np.pi*f*t) + noise  # Señal con ruido

# Aplicar filtro de media móvil
window_size = 11  # Tamaño de la ventana del filtro
filtered_signal = savgol_filter(signal, window_size, 1)  # Aplicar filtro

# Graficar la señal original y la señal filtrada
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# Señal original
ax[0].plot(t, signal)
ax[0].set_xlabel('Tiempo [s]')
ax[0].set_ylabel('Amplitud')
ax[0].set_title('Señal de coseno con ruido')

# Señal filtrada
ax[1].plot(t, filtered_signal)
ax[1].set_xlabel('Tiempo [s]')
ax[1].set_ylabel('Amplitud')
ax[1].set_title('Señal de coseno filtrada')

plt.tight_layout()
plt.show()
```

Uso del filtro de media movil para eliminar el ruido de una señal, demarcar tendencias

La función savgol_filter de la librería Scipy en Python aplica un filtro de suavizado de Savitzky-Golay a una señal dada. Este filtro es un tipo de filtro de media móvil que se utiliza comúnmente para suavizar señales que contienen ruido. A diferencia de un filtro de media móvil simple, el filtro de Savitzky-Golay utiliza un ajuste polinómico local para suavizar la señal, lo que permite una mejor preservación de las características importantes de la señal, como los bordes y las transiciones rápidas.

La función savgol_filter acepta tres argumentos principales: la señal de entrada que se va a suavizar, el tamaño de la ventana del filtro y el orden del ajuste polinómico. El tamaño de la ventana del filtro (window_size) especifica el número de puntos que se utilizarán en cada ajuste polinómico local, mientras que el orden (polyorder) especifica el orden del polinomio que se ajustará a los puntos dentro de cada ventana.

La función devuelve la señal suavizada después de aplicar el filtro de Savitzky-Golay. En resumen, la función savgol_filter es una herramienta útil para reducir el ruido en señales y preservar sus características importantes.
