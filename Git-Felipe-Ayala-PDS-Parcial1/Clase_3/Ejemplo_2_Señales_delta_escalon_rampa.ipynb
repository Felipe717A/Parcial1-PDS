# Ejemplo2. Multiplicacion de se√±ales muestreadas con el escalon unitario desplazado

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una secuencia de n√∫meros aleatorios entre -5 y 5
np.random.seed(1234) # Para reproducibilidad
random_seq = np.random.uniform(-5, 5, 50)

# Crear un desplazamiento aleatorio
random_shift = 10

# Crear un escal√≥n unitario desplazado
step_seq = np.zeros_like(random_seq)
step_seq[random_shift:] = 1

# Multiplicar las secuencias
mult_seq = random_seq * step_seq

# Graficar las secuencias originales y la multiplicaci√≥n
fig, ax = plt.subplots(3, 1, figsize=(8, 8))

# Secuencia aleatoria
ax[0].stem(random_seq)
ax[0].set_xlabel('√çndice')
ax[0].set_ylabel('Valor')
ax[0].set_title('Secuencia aleatoria')

# Escal√≥n unitario desplazado
ax[1].stem(step_seq)
ax[1].set_xlabel('√çndice')
ax[1].set_ylabel('Valor')
ax[1].set_title('Escal√≥n unitario desplazado')

# Multiplicaci√≥n
ax[2].stem(mult_seq)
ax[2].set_xlabel('√çndice')
ax[2].set_ylabel('Valor')
ax[2].set_title('Multiplicaci√≥n')

plt.tight_layout()
plt.show()
```
